**7.2** Given `A = [[2,1],[1,2]]`, compute determinant, inverse, and verify `A @ A_inv ≈ I`.

```python
import numpy as np
A=np.array([[2, 1], [1, 2]])
det_A=np.linalg.det(A)
A_inv=np.linalg.inv(A)
product=A@A_inv
print('determinant:', det_A)
print('inverse:\n',A_inv)
print('A@A_inv:\n',product)
print('A@A_inv is close to identity:', np.allclose(product, np.eye(2)))
```

**7.3** For `A` above, compute eigenvalues and eigenvectors, and verify `A @ v ≈ λ v`.

```python
import numpy as np
A=np.array([[2, 1], [1, 2]])
eigenvalues,eigenvectors=np.linalg.eig(A)
print('eigenvalues:', eigenvalues)
print('eigenvectors:\n', eigenvectors)
for i, lam in enumerate(eigenvalues):
    v=eigenvectors[:, i]
    print(f'chek for eigenvalue{lam}:', np.allclose(A @ v, lam * v))
```

**9.4 A/B Testing**: Variant A converts 48/500 users; Variant B converts 62/500. Compute conversion rates and the **absolute lift** (B−A). Simulate 10,000 bootstrap samples of the lift.

```python
import numpy as np
np.random.seed(0)
conversions_a = 48
users_a = 500
conversions_b = 62
users_b = 500
rate_a = conversions_a / users_a
rate_b = conversions_b / users_b
absolute_lift = rate_b - rate_a
samples = 10000
simulated_lifts = np.empty(samples)
for i in range(samples):
    sample_a = np.random.binomial(users_a, rate_a) / users_a
    sample_b = np.random.binomial(users_b, rate_b) / users_b
    simulated_lifts[i] = sample_b - sample_a
print('Conversion rate A:', rate_a)
print('Conversion rate B:', rate_b)
print('Absolute lift (B-A):', absolute_lift)
print('Bootstrap mean lift:', simulated_lifts.mean())
print('Bootstrap std of lift:', simulated_lifts.std())
```
